```python
import pandas as pd
import matplotlib.pyplot as plt

#Para graficar galaxias de un cg

# Supongamos que tu DataFrame se llama 'cg'
# Asegúrate de tener las columnas adecuadas en tu DataFrame

# Filtrar galaxias del primer grupo
grupo_1 = cg.head(4)

# Filtrar galaxias del segundo grupo
grupo_2 = cg.tail(4)

# Función para graficar proyecciones XY, YZ y XZ de un grupo
def graficar_proyecciones(axs, grupo, titulo, color):
    galaxia_host = grupo.iloc[0]
    
    # Ajustar posiciones para que la galaxia host sea (0, 0)
    x_adjusted = grupo['X'] - galaxia_host['X']
    y_adjusted = grupo['Y'] - galaxia_host['Y']
    z_adjusted = grupo['Z'] - galaxia_host['Z']

    axs[0].scatter(x_adjusted, y_adjusted, label=titulo, color=color, marker='*')
    axs[0].set_xlabel('X[kpc]')
    axs[0].set_ylabel('Y[kpc]')
    axs[0].set_title(f'XY Projection - {titulo}')
    axs[0].set_xlim(-2 * galaxia_host['R200c']-10, 2 * galaxia_host['R200c']+10)
    axs[0].set_ylim(-2 * galaxia_host['R200c']-10, 2 * galaxia_host['R200c']+10)

    axs[1].scatter(y_adjusted, z_adjusted, label=titulo, color=color, marker='*')
    axs[1].set_xlabel('Y[kpc]')
    axs[1].set_ylabel('Z[kpc]')
    axs[1].set_title(f'YZ Projection - {titulo}')
    axs[1].set_xlim(-2 * galaxia_host['R200c']-10, 2 * galaxia_host['R200c']+10)
    axs[1].set_ylim(-2 * galaxia_host['R200c']-10, 2 * galaxia_host['R200c']+10)

    axs[2].scatter(x_adjusted, z_adjusted, label=titulo, color=color, marker='*')
    axs[2].set_xlabel('X[kpc]')
    axs[2].set_ylabel('Z[kpc]')
    axs[2].set_title(f'XZ Projection - {titulo}')
    axs[2].set_xlim(-2 * galaxia_host['R200c']-10, 2 * galaxia_host['R200c']+10)
    axs[2].set_ylim(-2 * galaxia_host['R200c']-10, 2 * galaxia_host['R200c']+10)

    # Dibujar círculo centrado en la galaxia host con radio 2*R200c
    axs[0].add_patch(plt.Circle((0, 0), 2 * galaxia_host['R200c'], fill=False, color='black', linestyle='dashed'))
    axs[1].add_patch(plt.Circle((0, 0), 2 * galaxia_host['R200c'], fill=False, color='black', linestyle='dashed'))
    axs[2].add_patch(plt.Circle((0, 0), 2 * galaxia_host['R200c'], fill=False, color='black', linestyle='dashed'))

# Crear subgráficos para XY, YZ y XZ
fig, axs = plt.subplots(2, 3, figsize=(15, 10))

# Graficar proyecciones para el primer grupo
graficar_proyecciones(axs[0, :], grupo_1, 'MainHaloID: 12569477717', 'blue')

# Graficar proyecciones para el segundo grupo
graficar_proyecciones(axs[1, :], grupo_2, 'MainHaloID: 12587440177', 'red')

# Ajustar diseño y mostrar gráficos
plt.tight_layout()
plt.show()
```
